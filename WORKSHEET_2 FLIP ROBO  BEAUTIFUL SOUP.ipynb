# 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
and the code for the video from the link for the youtube video from the post

```python
# h1 class="entry-title"]  time[class="entry-time"] p div [class="entry-content"] div [class="ytp-menuitem-label"]
```

# IMPORTANT REQUIRED LIBRARIES

```python
from bs4 import BeautifulSoup
import requests
```

# FIRST WE HAVE TO GIVE REQUEST TO SERVER OF THE WEBSITE

```python
# AFTER GIVING REQUEST IF WE FIND THE RESPONSE BETWEEN  100-300 WE CAN ABLE TO WEB SCRAPP FROM THE WEBSITE.
page= requests.get('https://coreyms.com')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

# SCRAPPING HEADING OF THE WEBPAGE

```python
heading_page=soup.find('header',class_="entry-header")
heading_page                       
```

# SCRAPPING DATE OF THE WEBPAGE

```python
date_ofpage = soup.find('time',class_="entry-time")
date_ofpage
```

# SCRAPPING CONTENT OF THE PAGE

```python
content_ofpage =soup.find('div',class_="entry-content")
content_ofpage
```

# SCRAPPING URL LINK OF THE VIDEO IN WEBPAGE

```python
#dict, using the get function.

import requests
from bs4 import BeautifulSoup
# Make a request
page = requests.get("https://coreyms.com")
soup = BeautifulSoup(page.content, 'html.parser')

# Create top_items as empty list
all_links = []

# Extract and store in top_items according to instructions on the left
links = soup.select('a')
for ahref in links:
    text = ahref.text
    text = text.strip() if text is not None else ''

    href = ahref.get('href')
    href = href.strip() if href is not None else ''
    all_links.append({"href": href, "text": text})

print(all_links)
```
